## Plotting originial signal and the signal with noise
np.random.randn(len(t)) random data from 0 to 1 from standard normal distribution


```python
dt= 0.001
t=np.arange(0,1,dt)
f=np.sin(2*np.pi*50*t) + np.sin(2*np.pi*120*t)+ np.sin(2*np.pi*235*t)

f_noise=f+2.5*np.random.randn(len(t))

## time plot

plt.plot(t,f,color='c', label='Original Signal')
plt.plot(t,f_noise,color='r', label='Signal with noise')
plt.legend()
plt.xlim(t[0],t[-1])
plt.xlabel('time')
plt.ylabel('amplitude')


```

## Power sectral density calculation 

Unit : V^2/Hz

```python
n=len(t)
fhat=np.fft.fft(f_noise,n) ## FFT of the noisy signal 
PSD=fhat*np.conj(fhat)/n ## power spectral density calculation

freq=(1/(dt*n))*np.arange(n)
L=np.arange(1,np.floor(n/2),dtype='int')

# Frequency plot

plt.plot(freq[L],PSD[L],color='c', label='Original Signal')
plt.legend()
plt.xlim(freq[0],freq[-1])
plt.xlabel('frequency (Hz)')


```

```python
## filtering the signal
filter= PSD>50
PSD_filt=PSD*filter
fhat_filt= fhat*filter
ffilt=np.fft.ifft(fhat_filt)


## time plot
plt.plot(t,ffilt,color='r', label='Signal with noise')
plt.plot(t,f,color='c', label='Original Signal')

plt.legend()
plt.xlim(t[0],t[-1])
plt.xlabel('time')
plt.ylabel('amplitude')




```
