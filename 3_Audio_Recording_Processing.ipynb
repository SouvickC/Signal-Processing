```python
import sounddevice as sd
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf  # librosa uses this internally
import IPython.display as ipd

# Record audio
fs = 16000   
duration = 5 
print(f"Recording for {duration} secs ...")
audio = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype='float32')
sd.wait()
print("Recording finished.")

# saving the recording file
sf.write(r"D:\CS\Signal-Processing\Sample Data\recording.wav", audio, fs)


```

```python
# Load the audio file
y, sr = librosa.load(r"D:\CS\Signal-Processing\Sample Data\recording.wav", sr=2250)  # sr=None keeps original sample rate

# Playback

ipd.Audio(y, rate=sr) 
```
